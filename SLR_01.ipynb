Simple Linear Regression With scikit-learn

```
import numpy as np
import matplotlib.pyplot as plt

from sklearn.linear_model import LinearRegression

x = np.array([8, 18, 23, 36, 47, 50, 36, 26, 17]).reshape((-1, 1))
y = np.array([5, 20, 14, 32, 22, 38, 30, 17, 24])
```

```
plt.scatter(x, y, c=y, cmap ='flag')
plt.xlabel('Visitors')
plt.ylabel('Sales (Rs in Thousands)')
plt.show()
```

```
slr = LinearRegression()
slr.fit(x, y)
y_predict = slr.predict(x)
```

```
print(y_predict)
```

```
plt.scatter(x, y, c=y, cmap ='flag')
plt.plot(x, y_predict, color='blue', linewidth=3)

plt.xlabel('Visitors')
plt.ylabel('Sales (Rs in Thousands)')
plt.show()
```
